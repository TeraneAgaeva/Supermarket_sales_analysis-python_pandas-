9.  Створимо нову колонку в даних Rating Category, яка визначає категорію рейтингу: високий, середній чи низький за наступним правилом:

якщо рейтинг вищий або рівний 8 - то він високий (High)
якщо рейтинг від 6 до 8 не включаючи - то він середній (Medium)
інакше - низький (Low).

```python
def Category (rating):
  if rating >= 8:
        return 'High'
  elif rating >= 6 and rating < 8:
        return 'Medium'
  else:
        return 'Low'
```

```python
df['Rating Category'] = df['Rating'].apply(Category)
```

Виведемо, який відсоток даних становить кожне зі значень категорій рейтингу:
